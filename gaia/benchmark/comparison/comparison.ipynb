```python
import pandas as pd
```

```python
data_path: str = 'quarter_second'
```

```python
comparison_df = pd.read_csv(f'{data_path}/data.csv')
comparison_df
```

```python
comparison_df['ampere'] = comparison_df['Reading']
volt = 20
comparison_df['watt'] = comparison_df['ampere'] * volt
comparison_df['watt_norm'] = comparison_df['watt'] / comparison_df['watt'].max()
```

```python
# comparison_df['watt_norm'][comparison_df['watt_norm'] > 0.6][6500:].plot.line()
# comparison_df['watt_norm'][comparison_df['watt_norm'] > 0.4].dropna().plot.line()
comparison_df['watt_norm'][comparison_df['watt_norm'] > 0.4].rolling(500).mean().dropna().plot.line()
# comparison_df['watt_norm'][comparison_df['watt_norm'] > 0.65][5000:]
```

```python
greem_df = pd.read_csv(f'{data_path}/emissions.csv')
greem_df.head()
```

```python
greem_df['energy_consumed_norm'] = greem_df['energy_consumed'] / greem_df['energy_consumed'].max()
```
