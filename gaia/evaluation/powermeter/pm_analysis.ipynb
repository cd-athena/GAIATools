```python
df['project_name'] = df['project_name'].apply(lambda x: x.split('_')[0].removesuffix('s'))
```

```python
agg = df.groupby(df['project_name']).aggregate({'energy_consumed': 'mean'})
```

```python
agg.sort_values(by='project_name')
```
