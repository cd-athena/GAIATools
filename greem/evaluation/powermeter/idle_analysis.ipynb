```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pm_df = pd.read_csv('idle_powermeter_measurements_20_3.csv')
pm_df['watt'] = pm_df['Reading'] * 20
pm_df['watt_norm'] = pm_df['watt'] / pm_df['watt'].max()
pm_df.head(3)
```

```python
pm_df.describe()
```

```python
sns.lineplot(pm_df, x=pm_df.index, y='watt_norm')
```

```python
small_df = pm_df.where(pm_df['watt_norm'] < 0.55)
sns.lineplot(small_df, x=small_df.index, y='watt_norm')
```
